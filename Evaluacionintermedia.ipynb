```python
# 3
df.columns = df.columns.str.lower() # modifico sobre el df directamente
```

```python
# forma corregida y mas correcta (try&except)
def cambio(num):
    try:
        return float(num.replace(",", "."))
    except:
        return np.nan
lista_columnas =['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
for col in lista_columnas:
    df[col] = df[col].apply(cambio)
```

```python
# lo que yo había hecho
lista_col = ['danceability', 'energy', 'key', 'loudness', 'speechiness', 'acousticness', 'instrumentalness']
def decimales(df, col): # funcion para cambiar comas por puntos y pasar a float
    for col in lista_col:
        df[col] = df[col].str.replace(',', '.').astype(float) #reemplazo y paso a float
    return df
```

```python
df = decimales(df, lista_col) # llamo a la función
```

```python
df.info() #reviso que los cambios están ok
```

```python
def mayus(df, columnas): #Creo una función para pasar de mayuscula a minuscula y luego iterar por las tres columnas
    for col in columnas:
        df[col] = df[col].str.lower()
    return df
```

```python
columnas_mayus = ['track', 'title', 'album'] #lista con las columnas que quiero iterar
df = mayus(df, columnas_mayus) # llamo a la función e incluyo la lista de las columnas a cambiar mayus por minus
```

```python
df['track'] # reviso que las canciones están en minus
```

```python
df['artist'] = df['artist'].str.replace('$', 's') # cambio el simbolo por s 
```

```python
def dolar(artistas): # me lo piden en función
    return artistas.str.replace('$', 's')
```

```python
df['artist'] = dolar(df['artist']) # llamo a la función para cambiar simbolo dolar en columna artista
```

```python
df['artist'] #reviso que se ha hecho el cambio
```

```python
# 4 filtrar datos
unicos = df['album_type'].unique() #miro los valores únicos y almaceno en una variable para crear los dataframe de después
```

```python
# creo los tres df accediendo a los datos de cada valor único de album type por índice
df_album_album = df[df['album_type'] == unicos[0]]
df_album_single = df[df['album_type'] == unicos[1]]
df_album_compilation = df[df['album_type'] == unicos[2]]
```

```python
# media y desviacion estándar para columnas ordenado por media
# no es necesario el reset index, lo incluimos para ver indice
columnas = ['danceability', 'acousticness', 'speechiness']
df_metricas1 = df_album_album.groupby('artist')['danceability'].agg(['mean', 'std']).reset_index().sort_values(by='mean')
df_metricas2 = df_album_album.groupby('artist')['acousticness'].agg(['mean', 'std']).reset_index().sort_values(by='mean')
df_metricas3 = df_album_album.groupby('artist')['speechiness'].agg(['mean', 'std']).reset_index().sort_values(by='mean')
```

```python
# agrupar por artista y calcular la media de visitas y likes

df_views = df.groupby('artist')['views'].mean().sort_values(ascending=False)
df_likes = df.groupby('artist')['likes'].mean().sort_values(ascending=False)

# 10 artistas + visitas y likes
print(df_views.head(10))
print(df_likes.head(10))

# Ed Sheeran, Charlie Puth, Luis Fonsi y Justin Bieber están en ambos top, en el 40% de los casos si hay una relación entre ambos kpis
```
